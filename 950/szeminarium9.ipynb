```python
from scipy.optimize import fsolve
```

```python
def rbc(variables):
    K = variables[0]
    L = variables[1]
    # Kamatlab 1=ß*(1+r)
    r = 1/0.97-1
    # Tőkekínálat rk-d=r
    rk = r+0.02
    Y = 1.2*K**0.37*L**0.63
    w = 0.63*Y/L
    C = (0.65*w/L**0.21)**(1/0.35)
    I = K*0.02
    # Árupiaci egyensúly Y = C+I
    e1 = Y-C-I
    # Tőkekereslet rk=0.37*Y/K
    e2 = rk-0.37*Y/K
    return [e1, e2]
```

```python
Keq, Leq = fsolve(rbc, [100,2])
```

```python
import numpy
Kin = list(numpy.linspace(30,Keq,99))
```

```python
Lin = [Leq]*100
```

```python
IN = Kin+Lin
```

```python
def rbcdin(KLertekek):
    K = numpy.asarray([30]+list(KLertekek[0:99])+[Keq])
    L = numpy.asarray(list(KLertekek[99:])+[Leq])
    e = []
    # 99 Euler egyenlet
    for t in range(99):
        # u_ct=u_ct+1*ß*(1+r_t+1)
        Yt = 1.2*K[t]**0.37*L[t]**0.63
        Ytp1 = 1.2*K[t+1]**0.37*L[t+1]**0.63
        It = K[t+1]-K[t]*(1-0.02)
        Itp1 = K[t+2]-K[t+1]*(1-0.02)
        Ct = Yt-It
        Ctp1 = Ytp1-Itp1
        rktp1 = 0.37*Ytp1/K[t+1]
        rtp1 = rktp1-0.02
        e.append(Ct**(-0.35)-Ctp1**(-0.35)*0.97*(1+rtp1))
    for t in range(100):
        # Munkakínálat
        Yt = 1.2*K[t]**0.37*L[t]**0.63
        It = K[t+1]-K[t]*(1-0.02)
        Ct = Yt-It
        wt = 0.63*Yt/L[t]
        e.append(wt*0.65*Ct**(-0.35)-L[t]**0.21)
    return e
```

```python
fsolve(rbcdin, IN)
```
