```python
from scipy.optimize import fsolve
```

```python
def rbc(var):
    K = var[0]
    L = var[1]
    # Euler 1=ß*(1+r)
    r = 1/0.97-1
    # Tőkekínálat rk-d=r
    rk = r+0.02
    # Termelési függvény
    Y = 1.2*K**0.37*L**0.63
    # Beruházás
    I = 0.02*K
    # Munkakereslet
    w = 0.63*Y/L
    # Munkakínálat
    C = (w*0.65/L**0.21)**(1/0.35)
    # Árupiaci egyensúly
    e1 = Y-C-I
    # Tőkekereslet
    e2 = rk-0.37*Y/K
    return [e1, e2]
```

```python
Keq, Leq = fsolve(rbc, [100, 2])
```

```python
import numpy
```

```python
Kin = list(numpy.linspace(30,Keq,99))
```

```python
Lin = [Leq]*100
```

```python
IN = Kin+Lin
```

```python
def rbcdin(KLertekek):
    K = numpy.asarray([30]+list(KLertekek[0:99])+[Keq])
    L = numpy.asarray(list(KLertekek[99:])+[Leq])
    e = []
    for t in range(99):
        Yt=1.2*K[t]**0.37*L[t]**0.63
        Ytp1 = 1.2*K[t+1]**0.37*L[t+1]**0.63
        It = K[t+1]-K[t]*(1-0.02)
        Itp1 = K[t+2]-K[t+1]*(1-0.02)
        Ct = Yt-It
        Ctp1 = Ytp1-Itp1
        rktp1 = 0.37*Ytp1/K[t+1] # tőkekereslet
        rtp1 = rktp1-0.02 # tőkekínálat
        e.append(Ct**(-0.35)-Ctp1**(-0.35)*0.97*(1+rtp1))
    for t in range(100):
        Yt=1.2*K[t]**0.37*L[t]**0.63
        It = K[t+1]-K[t]*(1-0.02)
        Ct = Yt-It
        wt = L[t]**0.21/0.65*Ct**0.35 # Munkakínálat
        e.append(wt-0.63*Yt/L[t])
    return e
```

```python
fsolve(rbcdin, IN)
```
